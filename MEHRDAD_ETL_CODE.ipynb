```python
import pandas as pd
from sqlalchemy import create_engine
import json
```

```python
csv_CAvideos = '../Resources/CAvideos.csv'
csv_GBvideos = '../Resources/GBvideos.csv'
csv_USvideos = '../Resources/USvideos.csv'


CAvideos_data_df = pd.read_csv(csv_CAvideos)
GBvideos_data_df = pd.read_csv(csv_GBvideos)
USvideos_data_df = pd.read_csv(csv_USvideos)
```

```python
us_json_file = '../Resources/US_category_id.json'
US_category_json_df = pd.read_json(us_json_file)
US_category_json_df.head()
```

```python
USvideos_df = USvideos_data_df[['video_id', 'trending_date', 'title', 'channel_title', 'category_id',
       'publish_time', 'tags', 'views', 'likes', 'dislikes', 'comment_count',
       'thumbnail_link', 'comments_disabled', 'ratings_disabled',
       'video_error_or_removed', 'description']].copy()
```

```python
USvideos_df.insert(1,'country','US')
USvideos_df.head()
```

```python
top_trending_vidoes=USvideos_df[['video_id','title', 'views', 'trending_date', 'category_id', 'country']]
top_trending_vidoes.drop_duplicates('video_id', inplace=True)
top_trending_vidoes = top_trending_vidoes.sort_values(by= 'views', ascending=False )
top_trending_vidoes = top_trending_vidoes.reset_index(drop=True)
top_trending_vidoes.head(10)
```

```python
top_liked_vidoes=USvideos_df[['video_id','title', 'likes', 'category_id', 'country']]
top_liked_vidoes.drop_duplicates('video_id', inplace=True)
top_liked_vidoes = top_liked_vidoes.sort_values(by= 'likes', ascending=False )
top_liked_vidoes = top_liked_vidoes.reset_index(drop=True)
top_liked_vidoes.head(10)
```

```python
Highest_comment_count_vidoes=USvideos_df[['video_id','title','comment_count', 'country']]
Highest_comment_count_vidoes.drop_duplicates('video_id', inplace=True)
Highest_comment_count_vidoes = Highest_comment_count_vidoes.sort_values(by= 'comment_count', ascending=False )
Highest_comment_count_vidoes = Highest_comment_count_vidoes.reset_index(drop=True)
Highest_comment_count_vidoes.head(10)
```
