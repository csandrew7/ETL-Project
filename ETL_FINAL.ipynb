```python
# DEPENDENCIES
import pandas as pd
from sqlalchemy import create_engine
import pymongo
import json
```

```python
# RESOURCES FILES AND CONVERTING INTO DATAFRAME
csv_CAvideos = 'Resources/CAvideos.csv'
csv_GBvideos = 'Resources/GBvideos.csv'
csv_USvideos = 'Resources/USvideos.csv'

CAvideos_data_df = pd.read_csv(csv_CAvideos)
GBvideos_data_df = pd.read_csv(csv_GBvideos)
USvideos_data_df = pd.read_csv(csv_USvideos)
```

### CANADA TRENDING VIDEOS

```python
CAvideos_data_df.head()
```

```python
# CLEANING TABLES TO GIVE COLUMNS NEEDED
clean_cavideos_df = CAvideos_data_df[['video_id', 'title', 'channel_title', 'category_id', 'views', 'likes', 'dislikes', 'comment_count']]
clean_cavideos_df.insert(1,'country', 'CA')
clean_cavideos_df
```

```python
# READ THE JSON FILE
ca_json_file = "Resources/CA_category_id.json"
```

```python
# CONVERT JSON FILE INTO A PANDAS DATAFRAME
ca_json_df = pd.read_json(ca_json_file)
ca_json_df.head()
```

```python
list_countries=["Resources/CA_category_id.json"]
category = []

for country in list_countries:
    json_data = json.load(open(country,'r'))
    for data in json_data['items']:
        category.append([data['id'], data['snippet']['title']])

category
```

```python
# USE THE MAP FUNCTION TO ASSIGN CATEGORY ID WITH A CATEGORY NAME BASED ON THE JSON FILE
categoryCA_map={1:'Film & Animation', 2:'Autos & Vehicles', 10:'Music', 15:'Pets & Animals', 17:'Sports',18: 'Short Movies',
               19:'Travel & Events', 20:'Gaming',21: 'Videoblogging',22: 'People & Blogs',23: 'Comedy',24: 'Entertainment',
               25: 'News & Politics',26: 'Howto & Style',27: 'Education', 28: 'Science & Technology',29: 'Nonprofits & Activism',30: 'Movies', 
               31: 'Anime/Animation', 32: 'Action/Adventure',33: 'Classics', 34:'Comedy', 35: 'Documentary', 36: 'Drama',
               37: 'Family', 38: 'Foreign', 39: 'Horror', 40: 'Sci-Fi/Fantasy', 41: 'Thriller', 42: 'Shorts', 43: 'Shows', 44: 'Trailers'}

clean_cavideos_df['category_name'] = clean_cavideos_df['category_id'].map(categoryCA_map)
clean_cavideos_df.head()

# clean_cavideos_df = clean_cavideos_df[['video_id', 'country', 'title', 'channel_title', 'category_id', 'category_name', 'views', 'likes', 'dislikes', 'comment_count']]
# clean_cavideos_df
```

```python
# TOP 10 TRENDNING VIDEOS
top10_videos_cols = ['video_id', 'title', 'channel_title', 'views',  'category_id', 'category_name', 'country']
top10_videos_ca = clean_cavideos_df[top10_videos_cols].copy()
top10_videos_ca.drop_duplicates('video_id', inplace = True)
top10_videos_ca = top10_videos_ca.sort_values(by='views', ascending=False)
top10_videos_ca = top10_videos_ca.reset_index(drop=True)
top10_videos_ca.head(10)
```

```python
# TOP 10 HIGHEST LIKES VIDEOS
top_liked_videos = clean_cavideos_df[['country', 'video_id', 'title', 'channel_title', 'views','likes', 'category_id','category_name']].sort_values('likes', ascending=False)
top_liked_videos.drop_duplicates('video_id', inplace=True)
top_liked_videos.reset_index(drop=True)
top_liked_videos.head(10)
```

```python
# TOP 10 HIGHEST COMMENT COUNT VIDEOS
highest_comment_count = clean_cavideos_df[['country','video_id', 'title', 'channel_title', 'comment_count']].sort_values('comment_count', ascending=False)
highest_comment_count.drop_duplicates('video_id', inplace=True)
highest_comment_count.reset_index(drop=True)
highest_comment_count.head(10)
```

```python
# HOW MANY VIDEOS FALL INTO EACH CATEGORY
ca_category_counts = clean_cavideos_df.groupby(['category_name']).size()
ca_category_counts = clean_cavideos_df.groupby(['category_name']).size().reset_index(name='counts')
ca_category_counts
```

```python
GB_json_file = "Resources/GB_category_id.json"
GB_json_df=pd.read_json(GB_json_file)
GB_json_df.head()
```

```python
list_countries = ["Resources/GB_category_id.json"]
category = []
for country in list_countries:
    json_data = json.load(open(country,'r'))
    for data in json_data['items']:
        category.append([data['id'],data['snippet']['title']])

category
```

```python
gb_cols = ['video_id', 'trending_date', 'title', 'channel_title', 'category_id',
       'publish_time', 'tags', 'views', 'likes', 'dislikes', 'comment_count',
       'thumbnail_link', 'comments_disabled', 'ratings_disabled',
       'video_error_or_removed', 'description']
GB_df = GBvideos_data_df[gb_cols].copy()
GB_df.insert(1,'country','GB')
GB_df.head()
```

```python
GB_category_map={1: 'Film & Animation', 2: 'Autos & Vehicles', 10: 'Music', 15: 'Pets & Animals', 17: 'Sports',
                 18: 'Short Movies', 19: 'Travel & Events', 20: 'Gaming', 21: 'Videoblogging',22: 'People & Blogs'
                 ,23: 'Comedy',24: 'Entertainment',25: 'News & Politics',26: 'Howto & Style',27:'Education',
                  28: 'Science & Technology',29: 'Nonprofits & Activism',30: 'Movies',31: 'Anime/Animation',
                  32: 'Action/Adventure',33: 'Classics',34: 'Comedy',35: 'Documentary',36: 'Drama',37: 'Family',
                 38: 'Foreign',39: 'Horror', 40: 'Sci-Fi/Fantasy',41: 'Thriller',42: 'Shorts',43: 'Shows',44: 'Trailers'}

GB_df['category_name']= GB_df['category_id'].map(GB_category_map)
GB_df.head()
```

```python
top_10_trending_cols = ["video_id", "title", "views", "category_id", "country"]
top_10_videos = GB_df[top_10_trending_cols].copy()
top_10_videos.drop_duplicates("video_id", inplace=True)
top_10_videos.set_index("video_id", inplace=False)
top10 = top_10_videos.sort_values(by='views', ascending=False).head(10)

top10
```

```python
top_viewed_videos_cols = ["video_id", "title", "likes", "trending_date", "category_id", "country"]
top_viewed_videos = GB_df[top_viewed_videos_cols].copy()
top_viewed_videos.drop_duplicates("video_id", inplace=True)
top_viewed_videos.set_index("video_id", inplace=False)
most_views = top_viewed_videos.sort_values(by='likes', ascending=False).head(10)
most_views
```

```python
comments_cols = ["video_id", "title", "comment_count", "category_id","country"]
top_comments = GB_df[comments_cols].copy()
top_comments.drop_duplicates("video_id", inplace=True)
top_comments.set_index("video_id", inplace=False)
top_comments = top_comments.sort_values(by='comment_count', ascending=False).head(10)
top_comments
```
